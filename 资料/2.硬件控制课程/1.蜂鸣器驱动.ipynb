```python
#调用GPIO底层库文件
import RPi.GPIO as GPIO
import time
```

```python
buzzer = 8  #蜂鸣器引脚开关，具体建议参考硬件参考手册
```

```python
#设置GPIO口为BCM编码方式
GPIO.setmode(GPIO.BCM)    
GPIO.setup(buzzer,GPIO.OUT) 
```

```python
def whistle():   
    GPIO.output(buzzer, GPIO.LOW)
    time.sleep(0.1)
    GPIO.output(buzzer, GPIO.HIGH)
    time.sleep(0.001)       
```

```python
whistle() #执行函数
```
