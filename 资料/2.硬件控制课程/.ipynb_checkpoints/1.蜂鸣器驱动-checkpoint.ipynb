```python
#导入Raspblock驱动库
from Raspblock import Raspblock
robot = Raspblock()      #获取对象
```

```python
robot.Buzzer_control(1)  #蜂鸣器开
```

```python
robot.Buzzer_control(0)  #蜂鸣器关
```

```python
del robot                #使用完成对象记住释放掉对象，不然下一个程序使用这个对象模块会被占用，导致无法使用
```
